# Third: Data Cleaning

```python
# Frequency tables for categorical features
for column in data.select_dtypes(include=['object']).columns:
    counts, bins = np.histogram(pd.crosstab(index=data[column], columns='% observations', normalize='columns'))
    plt.hist(bins[:-1], bins, weights=counts)
   # display(np.hist(pd.crosstab(index=data[column], columns='% observations', normalize='columns')))
    
# Histograms for each numeric feature
display(data.describe())
%matplotlib.inline
hist = data.hist(bins=30, sharey=True,figsize=(10,10))
```
